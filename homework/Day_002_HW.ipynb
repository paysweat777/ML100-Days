```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
print("rows:"+ str(app_train.shape[0]) + "\n"+ "columns:" + str(app_train.shape[1]))
```

#### 列出所有欄位

```python
for col in app_train.columns.values:
    print(col)
```

```python
# 完整顯示摘要,並顯示非空值列數
app_train.info(verbose=True,null_counts=True)
```

#### 截取部分資料

```python
# 以標籤取部分資料用loc
app_train.loc[:10,"FLAG_DOCUMENT_2":"FLAG_DOCUMENT_10"]
```

```python
#  以索引取部分資料用iloc
app_train.iloc[3:10,0:5]
```
