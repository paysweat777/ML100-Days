# 作業 : (Kaggle)房價預測

```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
df['1stFlrSF'].describe()
```

```python
sns.kdeplot(df['1stFlrSF'])
plt.show()
plt.boxplot(df['1stFlrSF'])
plt.show()
```

```python
# 計算離群值的上下界
def bound(x):
    q75, q25 = np.percentile(x, [75 ,25])
    iqr = q75 - q25
    upperbound = q75 + 1.5*iqr
    lowerbound = q25 - 1.5*iqr
    limit = (lowerbound,upperbound)
    return limit
```

```python
limit = bound(df['1stFlrSF'])
limit
```

```python
from scipy import stats
print('下界的百分位數',stats.percentileofscore(df['1stFlrSF'], limit[0]))
print('上界的百分位數',stats.percentileofscore(df['1stFlrSF'], limit[1]))
```

```python
import copy
df0 =copy.deepcopy(df)
df1 =copy.deepcopy(df)
train_Y0 = copy.deepcopy(train_Y)
train_Y1 = copy.deepcopy(train_Y)
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
"""
Your Code Here
"""
df0['1stFlrSF'] = df0['1stFlrSF'].clip(limit[0],limit[1])
sns.regplot(x = df0['1stFlrSF'], y=train_Y0)
plt.show()

# 做線性迴歸, 觀察分數
train_X0 = MMEncoder.fit_transform(df0)
estimator = LinearRegression()
cross_val_score(estimator, train_X0, train_Y0, cv=5).mean()
```

> 限制上下限，分數有提升

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""
keep_indexs1 = (df1['1stFlrSF']>limit[0]) & (df1['1stFlrSF']<limit[1])
df1 = df1[keep_indexs1]
train_Y1 = train_Y1[keep_indexs1]
sns.regplot(x = df1['1stFlrSF'], y = train_Y1)
plt.show()

# 做線性迴歸, 觀察分數
train_X1 = MMEncoder.fit_transform(df1)
estimator = LinearRegression()
cross_val_score(estimator, train_X1, train_Y1, cv=5).mean()
```
