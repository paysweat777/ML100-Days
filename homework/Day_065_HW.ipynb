# 作業:
嘗試調整參數:  
sg:sg=1表示採用skip-gram,sg=0 表示採用cbow  
window:能往左往右看幾個字的意思 

```python
import gensim, logging
from gensim.models import word2vec
```

```python
sentences = [['I am a hero', 'sentence'], ['She is a teacher', 'sentence']] 

# train word2vec on the two sentences  
# CBOW
cbow_model = word2vec.Word2Vec( sentences, size=300, min_count=1 , sg=0, window=5 )
# sg=0 表示COBW, sg=1 表示skip-gram
```

```python
print(cbow_model)
```

```python
cbow_model.similarity('I am a hero','She is a teacher')
```

```python
cbow_model["I am a hero"]
```

```python
# skip-gram
skipgram_model = word2vec.Word2Vec( sentences, size=300, min_count=1 , sg=1, window=5 )
# sg=0 表示COBW, sg=1 表示skip-gram
print(skipgram_model)
```

```python
skipgram_model.similarity('I am a hero','She is a teacher')
```

```python
skipgram_model["I am a hero"]
```

> 資料太少可能導致CBOW和Skip-gram看不出差異
