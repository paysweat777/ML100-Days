# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def relu(x):
    return np.maximum(0,x)

def drelu(x):
    return list(map(lambda x:1 if x>0 else 0,x))
```

```python
x = plt.linspace(-10,10,100)
plt.plot(x, relu(x), 'b', label='linspace(-10,10,100)')
plt.grid()
plt.title('Relu function')
 
# create the graph
plt.show()
```

```python
plt.plot(x, drelu(x), 'b', label='linspace(-10,10,100)')
plt.grid()
plt.title('Relu Derivative')
 
# create the graph
plt.show()
```
